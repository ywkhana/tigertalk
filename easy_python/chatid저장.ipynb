```python
import pandas as pd
df = pd.read_excel('talkdata.xlsx') 
```

```python
def isEnglishOrKorean(input_s):
    k_count = 0
    e_count = 0
    for c in input_s:
        if ord('ê°€') <= ord(c) <= ord('í£') or ord('ã„±') <= ord(c) <= ord('ã…') or ord('ã…') <= ord(c) <= ord('ã…£') :
            k_count+=1
        elif ord('a') <= ord(c.lower()) <= ord('z'):
            e_count+=1
    if k_count == 0 and e_count == 0:
        return "others"
    elif k_count>0:
        return "kor"
    else:
        return "eng"
```

```python
#----------ì–¸ì–´ Series ìƒì„±
lang = []
for i in df['plainText']:
    lang.append(isEnglishOrKorean(i))
df['language'] = pd.DataFrame(lang)
```

```python
kor_chat = df[df['language']=='kor']
```

```python
kor_chat.to_csv('kor_chat.csv',encoding='utf-8')
```

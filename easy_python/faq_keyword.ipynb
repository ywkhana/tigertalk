```python
import pandas as pd
df = pd.read_csv('question.csv',index_col=[0])
```

```python
df.head()
```

```python
import re
import konlpy
from tqdm import tqdm
from konlpy.tag import Okt
from collections import Counter
```

```python
text = df['plainText'].tolist()
print(len(text))
```

```python
def extract_word(text):
    hangul = re.compile('[^A-Za-z가-힣]')
    result = hangul.sub(' ', text)
    return result
```

```python
df = df.reset_index(drop=True)
df.head(3)
```

```python
df['plainText'] = df['plainText'].apply(lambda x: extract_word(x))
```

```python
# 형태소 분석
okt = Okt()
words = " ".join(df['plainText'].tolist())
words = okt.nouns(words)
```

```python
# 한 글자 단어(조사..) 제거
remove_one_word = [x for x in words if len(x)>1 or x == "줌"]
len(remove_one_word)
```

```python
from collections import Counter
frequent = Counter(remove_one_word).most_common()
frequent
```

```python
with open('stopwords.txt','r') as f:
    list_file = f.readlines()
stopwords = list_file[0].split(",")
remove_stopwords = [x for x in remove_one_word if x not in stopwords]
len(remove_stopwords)
```

```python
word_list = Counter(remove_stopwords).most_common()
word_list
```

```python
import csv
with open ("faq_keyword.csv","w",newline='',encoding='euc-kr') as f:
    csvw = csv.writer(f)
    for v in word_list:
        csvw.writerow(v)
```

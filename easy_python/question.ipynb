```python
## 질문만 뽑아내기 코드 ##
import pandas as pd
# 한국어 데이터만 불러오기
kor_chat = pd.read_csv('kor_chat.csv',index_col=[0])
```

```python
kor_chat.head()
```

```python
# unique한 chatid 파일 한줄씩 읽기
f = open('chatid2.txt',mode='r',encoding='utf-8')
id = f.readlines()
id = list(map(lambda s: s.strip(), id))
```

```python
# 잘리는 내용 보이게 하기
pd.options.display.max_colwidth = None
```

```python
# 질문만 추출해서 데이터프레임으로 합치기
df = kor_chat.head(0)
for i in id:
    line = kor_chat.loc[kor_chat['chatID'] == i].head(1)
    line = line.loc[line['personType'] == 'user']
    df = pd.concat([df, line])
df
```

```python
from wordcloud import WordCloud
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("faq_keyword.csv",
                 names = ['text', 'count'],
                 encoding="CP949")
df
```

```python
dic = df.set_index("text").to_dict()["count"]
```

```python
wc = WordCloud(font_path = 'BMDOHYEON_ttf.ttf',
               background_color = 'white',
               width = 1000,
               height = 1000,
               max_words = 100,
               max_font_size = 300).generate_from_frequencies(dic)
```

```python
plt.figure()
plt.imshow(wc)
plt.axis('off')
```

```python
wc.to_file('wordcloud.jpg')
```
